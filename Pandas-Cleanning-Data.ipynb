```python
import pandas as pd
```

```python
df = pd.read_excel (r"C:\Users\Padka\OneDrive\Desktop\Learning Python\Customer Call List.xlsx")
df
```

```python
# Delete duplicate rows

df.drop_duplicates(inplace=True)
df
```

```python
# Delete a useless column

df.drop(columns = "Not_Useful_Column", inplace=True)
df
```

```python
# Strip unwanted characters from the "Last_Name" column

# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
# Strip unwanted characters from the "Phone_Number" column and format it as XXX-XXX-XXXX

# df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
# Split the "Address" column into three new columns: "Street Address", "State", and "Zip_code"

df[["Street Address","State","Zip_code"]] = df["Address"].str.split(',', n=2, expand=True)
df
```

```python
# Replace "Yes" with "Y" and "No" with "N" in the "Paying Customer" column

df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
# Replace "Yes" with "Y" and "No" with "N" in the "Do_Not_Contact" column

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df["Do_Not_Contact"]
```

```python
# Replace "N/a" and "NaN" with an empty string in the entire DataFrame

# df = df.replace('N/a','')
# df = df.replace('NaN','')
df = df.fillna('')
df
```

```python
# Delete rows where the "Do_Not_Contact" column has a value of "Y"

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)
df
```

```python
# Delete rows where the "Phone_Number" column has an empty string as a value

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)
df

# df.dropna(subset = "Phone_Number",inplace = True)
```

```python
# Delete the "Address" column since we have already split it into three new columns

df = df.drop(columns= "Address")
```

```python
# Reset the index of the DataFrame after deleting rows

df = df.reset_index(drop=True)
df
```
